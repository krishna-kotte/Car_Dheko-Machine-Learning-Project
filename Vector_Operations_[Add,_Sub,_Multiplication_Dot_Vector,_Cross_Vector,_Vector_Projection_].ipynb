```python
#import the libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#setup our visual theme using seaborn
sns.set()
```

```python
#plot a vector
plt.quiver(0,0,5,8)
plt.show
```

```python
plt.quiver(0,0,5,8, scale_units ='xy',angles ='xy',scale=1,color ='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
#Addition in vector
vector_1 = np.asarray([0,0,5,7])
vector_2 = np.asarray([0,0,3,3])

Add_vector = vector_1 + vector_2
print(Add_vector)
```

```python
#Addition visualization
plt.quiver(0,0,5,7,scale_units ='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,3,scale_units ='xy',angles='xy',scale=1,color='b')

plt.quiver(0,0,8,10,scale_units ='xy',angles='xy',scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
#Subtraction in vector
vector_1 = np.asarray([0,0,5,7])
vector_2 = np.asarray([0,0,3,3])

Sub_vector = vector_1 - vector_2
print(Add_vector)
```

```python
#Subtraction visualization
plt.quiver(0,0,5,7,scale_units ='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,3,scale_units ='xy',angles='xy',scale=1,color='b')

plt.quiver(0,0,2,4,scale_units ='xy',angles='xy',scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
#Subtraction visualization
plt.quiver(0,0,5,7,scale_units ='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-3,scale_units ='xy',angles='xy',scale=1,color='b')

plt.quiver(0,0,2,10,scale_units ='xy',angles='xy',scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
#Multiplication of two vectors
Vector_1 = np.asarray([0,0,3,5])
scalar_1 = 5

Mul_Vector = Vector_1 * scalar_1
print(Mul_Vector)
```

```python
#Multiplication visualization
plt.quiver(0,0,5,7,scale_units ='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,3,scale_units ='xy',angles='xy',scale=1,color='b')

plt.quiver(0,0,15,25,scale_units ='xy',angles='xy',scale=1,color='g')
plt.xlim(-30,30)
plt.ylim(-30,30)
plt.show()
```
